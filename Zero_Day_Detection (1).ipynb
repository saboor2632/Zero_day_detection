```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
import os
import numpy as np
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from collections import Counter
import time
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix, classification_report, silhouette_score

class ZeroDayDetector:
    def __init__(self):
        self.scaler = StandardScaler()
        self.label_encoder = LabelEncoder()
        self.pca = PCA(n_components=0.95)  # Preserve 95% of variance

    def preprocess_data(self, df):
        """Data Preprocessing Step"""
        print("Step 1: Data Preprocessing")

        # Handle missing and infinite values
        df = df.replace([np.inf, -np.inf], np.nan)
        df = df.dropna()

        # Separate features and labels
        X = df.drop(' Label', axis=1)
        y = df[' Label']

        # Encode labels
        y_encoded = self.label_encoder.fit_transform(y)

        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
        )

        # Scale features
        X_train_scaled = self.scaler.fit_transform(X_train)
        X_test_scaled = self.scaler.transform(X_test)

        return X_train_scaled, X_test_scaled, y_train, y_test, X.columns

    def feature_engineering(self, X_train_scaled, X_test_scaled, feature_names):
        """Feature Selection and Engineering Step"""
        print("\nStep 2: Feature Engineering")

        # Apply PCA
        X_train_pca = self.pca.fit_transform(X_train_scaled)
        X_test_pca = self.pca.transform(X_test_scaled)

        # Get feature importance
        components_df = pd.DataFrame(
            self.pca.components_.T,
            columns=[f'PC{i+1}' for i in range(self.pca.components_.shape[0])],
            index=feature_names
        )

        # Plot feature importance
        plt.figure(figsize=(10, 6))
        sns.heatmap(components_df.head(15), cmap='viridis')
        plt.title('Top 15 Features Contribution to Principal Components')
        plt.tight_layout()
        plt.show()

        return X_train_pca, X_test_pca

    def train_models(self, X_train, X_test, y_train, y_test):
        """Model Training and Evaluation"""
        print("\nStep 3: Model Training and Evaluation")

        models = {
            'One-Class SVM': OneClassSVM(kernel='rbf', nu=0.1),
            'Isolation Forest': IsolationForest(contamination=0.1, random_state=42),
            'K-Means': KMeans(n_clusters=len(np.unique(y_train)), random_state=42)
        }

        results = {}

        for name, model in models.items():
            print(f"\nTraining {name}...")

            # Train model
            model.fit(X_train)

            # Get predictions
            if name == 'K-Means':
                train_pred = model.predict(X_train)
                test_pred = model.predict(X_test)
            else:
                train_pred = model.predict(X_train)
                test_pred = model.predict(X_test)
                # Convert predictions to binary (1 for inliers, 0 for outliers)
                train_pred = np.where(train_pred == 1, 0, 1)
                test_pred = np.where(test_pred == 1, 0, 1)

            # Calculate metrics
            train_accuracy = accuracy_score(y_train, train_pred)
            test_accuracy = accuracy_score(y_test, test_pred)

            train_precision = precision_score(y_train, train_pred, average='weighted', zero_division=0)
            test_precision = precision_score(y_test, test_pred, average='weighted', zero_division=0)

            train_recall = recall_score(y_train, train_pred, average='weighted', zero_division=0)
            test_recall = recall_score(y_test, test_pred, average='weighted', zero_division=0)

            train_f1 = f1_score(y_train, train_pred, average='weighted', zero_division=0)
            test_f1 = f1_score(y_test, test_pred, average='weighted', zero_division=0)

            train_report = classification_report(y_train, train_pred)
            test_report = classification_report(y_test, test_pred)

            results[name] = {
                'train_pred': train_pred,
                'test_pred': test_pred,
                'train_report': train_report,
                'test_report': test_report,
                'train_accuracy': train_accuracy,
                'test_accuracy': test_accuracy,
                'train_precision': train_precision,
                'test_precision': test_precision,
                'train_recall': train_recall,
                'test_recall': test_recall,
                'train_f1': train_f1,
                'test_f1': test_f1
            }

            # Plot confusion matrix
            plt.figure(figsize=(10, 5))

            plt.subplot(1, 2, 1)
            sns.heatmap(confusion_matrix(y_train, train_pred),
                        annot=True, fmt='d', cmap='Blues')
            plt.title(f'{name} - Training Confusion Matrix')

            plt.subplot(1, 2, 2)
            sns.heatmap(confusion_matrix(y_test, test_pred),
                        annot=True, fmt='d', cmap='Blues')
            plt.title(f'{name} - Testing Confusion Matrix')

            plt.tight_layout()
            plt.show()

            print(f"{name} Metrics:")
            print(f"Train Accuracy:  {train_accuracy:.4f}, Precision: {train_precision:.4f}, Recall: {train_recall:.4f}, F1 Score: {train_f1:.4f}")
            print(f"Test  Accuracy:  {test_accuracy:.4f}, Precision: {test_precision:.4f}, Recall: {test_recall:.4f}, F1 Score: {test_f1:.4f}")
            print(f"\n{name} Training Report:")
            print(train_report)
            print(f"\n{name} Testing Report:")
            print(test_report)

        return results


    def visualize_results(self, X, y, predictions, title):
        """Visualization of Results"""
        from sklearn.manifold import TSNE

        # Apply t-SNE
        tsne = TSNE(n_components=2, random_state=42)
        X_tsne = tsne.fit_transform(X)

        # Plot results
        plt.figure(figsize=(12, 4))

        # Original labels
        plt.subplot(1, 2, 1)
        scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
        plt.title(f'{title} - True Labels')
        plt.colorbar(scatter)

        # Predicted labels
        plt.subplot(1, 2, 2)
        scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=predictions, cmap='viridis')
        plt.title(f'{title} - Predicted Labels')
        plt.colorbar(scatter)

        plt.tight_layout()
        plt.show()

def main():
    # Start timing
    start_time = time.time()

    # Initialize detector
    detector = ZeroDayDetector()

    # Load data
    print("Loading data...")
    df = pd.read_csv('combined_friday_samples2.csv')
    print("\nOriginal class distribution:")
    print(df[' Label'].value_counts())

    # Preprocess data
    X_train_scaled, X_test_scaled, y_train, y_test, feature_names = detector.preprocess_data(df)

    # Feature engineering
    X_train_pca, X_test_pca = detector.feature_engineering(X_train_scaled, X_test_scaled, feature_names)

    # Train and evaluate models
    results = detector.train_models(X_train_pca, X_test_pca, y_train, y_test)

    # Visualize results for each model
    for name, result in results.items():
        detector.visualize_results(X_test_pca, y_test, result['test_pred'], name)

    # Print execution time
    end_time = time.time()
    print(f"\nTotal execution time: {(end_time - start_time)/60:.2f} minutes")

if __name__ == "__main__":
    main()



```

```python
import pandas as pd
import os
import numpy as np
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler, LabelEncoder, RobustScaler
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE, ADASYN
from collections import Counter
import time
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix, classification_report, silhouette_score
from sklearn.metrics import classification_report, confusion_matrix, f1_score, precision_score, recall_score

class ZeroDayDetector:
    def __init__(self):
        self.scaler = StandardScaler()
        self.robust_scaler = RobustScaler()  # For outlier-resistant scaling
        self.label_encoder = LabelEncoder()
        self.pca = PCA(n_components=0.95)  # Preserve 95% of variance

    def preprocess_data(self, df):
        """Enhanced Data Preprocessing Step"""
        print("Step 1: Enhanced Data Preprocessing")

        # 1. Check for and handle duplicate rows
        duplicate_count = df.duplicated().sum()
        print(f"\nDuplicate rows found: {duplicate_count}")
        df = df.drop_duplicates()
        print(f"After removing duplicates: {len(df)} rows")

        # 2. Check for missing values
        missing_values = df.isnull().sum()
        print(f"\nColumns with missing values: {missing_values[missing_values > 0].index.tolist()}")

        # 3. Handle infinite values
        inf_count = np.isinf(df.select_dtypes(include=[np.number])).sum().sum()
        print(f"Infinite values found: {inf_count}")
        df = df.replace([np.inf, -np.inf], np.nan)

        # 4. Handle missing values (after replacing infinities with NaN)
        df = df.dropna()
        print(f"After handling missing/infinite values: {len(df)} rows")

        # 5. Check for and visualize class imbalance
        print("\nClass distribution before balancing:")
        class_dist = df[' Label'].value_counts()
        print(class_dist)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=class_dist.index, y=class_dist.values)
        plt.title('Class Distribution - Original Data')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # 6. Detect outliers
        self._detect_outliers(df)

        # 7. Separate features and labels
        X = df.drop(' Label', axis=1)
        y = df[' Label']

        # 8. Encode labels
        y_encoded = self.label_encoder.fit_transform(y)

        # 9. Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
        )

        # 10. Handle class imbalance with SMOTE
        print("\nApplying SMOTE to balance classes...")
        smote = SMOTE(random_state=42)
        X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

        print("Class distribution after SMOTE:")
        balanced_dist = pd.Series(y_train_resampled).value_counts()
        print(balanced_dist)

        # 11. Scale features (after SMOTE to prevent data leakage)
        print("\nScaling features...")
        X_train_scaled = self.scaler.fit_transform(X_train_resampled)
        X_test_scaled = self.scaler.transform(X_test)

        return X_train_scaled, X_test_scaled, y_train_resampled, y_test, X.columns

    def _detect_outliers(self, df):
        """Detect and visualize outliers in the dataset"""
        print("\nDetecting outliers...")

        # Select numeric columns only
        numeric_df = df.select_dtypes(include=[np.number]).drop(' Label', axis=1, errors='ignore')

        # Create boxplots for selected features (top 5 by variance)
        variances = numeric_df.var().sort_values(ascending=False)
        top_features = variances.index[:5].tolist()

        plt.figure(figsize=(15, 10))
        for i, feature in enumerate(top_features):
            plt.subplot(2, 3, i+1)
            sns.boxplot(x=df[' Label'], y=df[feature])
            plt.title(f'Outliers in {feature}')
            plt.xticks(rotation=45)

        plt.tight_layout()
        plt.show()

        # Use IsolationForest to detect outliers
        iso_forest = IsolationForest(contamination=0.05, random_state=42)
        outliers = iso_forest.fit_predict(numeric_df)
        outlier_count = (outliers == -1).sum()

        print(f"Potential outliers detected: {outlier_count} ({outlier_count/len(df)*100:.2f}%)")

        # Visualize outliers using 2D PCA
        pca = PCA(n_components=2)
        pca_result = pca.fit_transform(numeric_df)

        plt.figure(figsize=(10, 6))
        plt.scatter(pca_result[outliers == 1, 0], pca_result[outliers == 1, 1],
                   c='blue', s=5, label='Normal')
        plt.scatter(pca_result[outliers == -1, 0], pca_result[outliers == -1, 1],
                   c='red', s=5, label='Outlier')
        plt.title('PCA visualization of outliers')
        plt.legend()
        plt.show()

        return outliers

    def feature_engineering(self, X_train_scaled, X_test_scaled, feature_names):
        """Enhanced Feature Selection and Engineering Step"""
        print("\nStep 2: Enhanced Feature Engineering")

        # 1. Check feature correlations
        print("Analyzing feature correlations...")

        # Convert scaled arrays back to DataFrames for correlation analysis
        X_train_df = pd.DataFrame(X_train_scaled, columns=feature_names)

        # Plot correlation matrix
        plt.figure(figsize=(12, 10))
        corr_matrix = X_train_df.corr()
        mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
        sns.heatmap(corr_matrix, mask=mask, cmap='coolwarm', vmax=1, vmin=-1, center=0,
                   square=True, linewidths=.5, annot=False)
        plt.title('Feature Correlation Matrix')
        plt.tight_layout()
        plt.show()

        # 2. Apply PCA
        print("Applying PCA...")
        X_train_pca = self.pca.fit_transform(X_train_scaled)
        X_test_pca = self.pca.transform(X_test_scaled)

        # Get number of components and explained variance
        n_components = self.pca.n_components_
        explained_variance = self.pca.explained_variance_ratio_.sum()
        print(f"Number of PCA components: {n_components}")
        print(f"Explained variance: {explained_variance:.2%}")

        # 3. Plot explained variance ratio
        plt.figure(figsize=(10, 6))
        plt.plot(np.cumsum(self.pca.explained_variance_ratio_))
        plt.xlabel('Number of Components')
        plt.ylabel('Cumulative Explained Variance')
        plt.title('Explained Variance by PCA Components')
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        # 4. Get feature importance
        components_df = pd.DataFrame(
            self.pca.components_.T,
            columns=[f'PC{i+1}' for i in range(self.pca.components_.shape[0])],
            index=feature_names
        )

        # Calculate and sort feature importances
        feature_importance = np.abs(self.pca.components_).mean(axis=0)
        feature_importance = pd.Series(feature_importance, index=feature_names).sort_values(ascending=False)

        # Plot top 15 most important features
        plt.figure(figsize=(12, 6))
        feature_importance.head(15).plot(kind='bar')
        plt.title('Top 15 Most Important Features')
        plt.tight_layout()
        plt.show()

        # 5. Plot feature contribution heatmap
        plt.figure(figsize=(12, 8))
        sns.heatmap(components_df.loc[feature_importance.index[:15]].T, cmap='viridis', annot=False)
        plt.title('Top 15 Features Contribution to Principal Components')
        plt.tight_layout()
        plt.show()

        return X_train_pca, X_test_pca, feature_importance
    def train_models(self, X_train, X_test, y_train, y_test):
        """Model Training and Evaluation"""
        print("\nStep 3: Model Training and Evaluation")

        models = {
            'One-Class SVM': OneClassSVM(kernel='rbf', nu=0.1, gamma='scale'),
            'Isolation Forest': IsolationForest(contamination=0.1, random_state=42, n_jobs=-1),
            'K-Means': KMeans(n_clusters=len(np.unique(y_train)), random_state=42, n_init=10),
            # 'DBSCAN': DBSCAN(eps=0.5, min_samples=5)
        }

        results = {}

        for name, model in models.items():
            print(f"\nTraining {name}...")

            # Train model
            model.fit(X_train)

            # Get predictions
            if name == 'K-Means':
                train_pred = model.predict(X_train)
                test_pred = model.predict(X_test)
            elif name == 'DBSCAN':
                train_pred = model.fit_predict(X_train)
                test_model = DBSCAN(eps=0.5, min_samples=5)
                test_pred = test_model.fit_predict(X_test)
                train_pred = np.where(train_pred == -1, max(train_pred) + 1, train_pred)
                test_pred = np.where(test_pred == -1, max(test_pred) + 1, test_pred)
            else:
                train_pred = model.predict(X_train)
                test_pred = model.predict(X_test)
                train_pred = np.where(train_pred == 1, 0, 1)
                test_pred = np.where(test_pred == 1, 0, 1)

            # Calculate metrics
            train_report = classification_report(y_train, train_pred, zero_division=0)
            test_report = classification_report(y_test, test_pred, zero_division=0)

            # Additional test metrics
            test_f1 = f1_score(y_test, test_pred, zero_division=0)
            test_precision = precision_score(y_test, test_pred, zero_division=0)
            test_recall = recall_score(y_test, test_pred, zero_division=0)

            results[name] = {
                'train_pred': train_pred,
                'test_pred': test_pred,
                'train_report': train_report,
                'test_report': test_report,
                'test_f1': test_f1,
                'test_precision': test_precision,
                'test_recall': test_recall
            }

            # Plot confusion matrix with proper labels
            plt.figure(figsize=(12, 5))

            plt.subplot(1, 2, 1)
            cm_train = confusion_matrix(y_train, train_pred)
            sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues')
            plt.title(f'{name} - Training Confusion Matrix')

            plt.subplot(1, 2, 2)
            cm_test = confusion_matrix(y_test, test_pred)
            sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues')
            plt.title(f'{name} - Testing Confusion Matrix')

            plt.tight_layout()
            plt.show()

            print(f"\n{name} Training Report:")
            print(train_report)
            print(f"\n{name} Testing Report:")
            print(test_report)
            print(f"F1 Score (Test): {test_f1:.4f}")
            print(f"Precision (Test): {test_precision:.4f}")
            print(f"Recall (Test): {test_recall:.4f}")

        return results

    def visualize_results(self, X, y, predictions, title):
        """Visualization of Results"""
        from sklearn.manifold import TSNE

        # Apply t-SNE
        tsne = TSNE(n_components=2, random_state=42)
        X_tsne = tsne.fit_transform(X)

        # Plot results
        plt.figure(figsize=(12, 4))

        # Original labels
        plt.subplot(1, 2, 1)
        scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
        plt.title(f'{title} - True Labels')
        plt.colorbar(scatter)

        # Predicted labels
        plt.subplot(1, 2, 2)
        scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=predictions, cmap='viridis')
        plt.title(f'{title} - Predicted Labels')
        plt.colorbar(scatter)

        plt.tight_layout()
        plt.show()

def main():
    # Start timing
    start_time = time.time()

    # Initialize detector
    detector = ZeroDayDetector()

    # Load data
    print("Loading data...")
    df = pd.read_csv('combined_friday_samples2.csv')

    # Preprocess data
    X_train_scaled, X_test_scaled, y_train, y_test, feature_names = detector.preprocess_data(df)

    # Feature engineering
    X_train_pca, X_test_pca, feature_importance = detector.feature_engineering(X_train_scaled, X_test_scaled, feature_names)

    # Train and evaluate models
    results = detector.train_models(X_train_pca, X_test_pca, y_train, y_test)

    # Print execution time
    end_time = time.time()
    print(f"\nTotal execution time: {(end_time - start_time)/60:.2f} minutes")

    # Print most important features
    print("\nTop 10 most important features:")
    print(feature_importance.head(10))

if __name__ == "__main__":
    main()


```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import OneClassSVM
from sklearn.ensemble import IsolationForest
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
from collections import Counter
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Model
import time
import json
import os
from sklearn.preprocessing import LabelEncoder, RobustScaler
from imblearn.over_sampling import SMOTE, ADASYN
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report, silhouette_score
from tensorflow.keras.callbacks import EarlyStopping

class Autoencoder:
    def __init__(self, input_dim, encoding_dim=32):
        self.input_dim = input_dim
        self.encoding_dim = encoding_dim
        self.autoencoder = self._build_autoencoder()

    def _build_autoencoder(self):
        # Encoder
        input_layer = Input(shape=(self.input_dim,))
        encoder = Dense(self.input_dim // 2, activation='relu')(input_layer)
        encoder = Dense(self.encoding_dim, activation='relu')(encoder)

        # Decoder
        decoder = Dense(self.input_dim // 2, activation='relu')(encoder)
        decoder = Dense(self.input_dim, activation='sigmoid')(decoder)

        # Autoencoder model
        autoencoder = Model(input_layer, decoder)
        autoencoder.compile(optimizer='adam', loss='mse')
        return autoencoder

    def fit(self, X_train, epochs=100, batch_size=32, validation_split=0.1):
        early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
        self.autoencoder.fit(
            X_train, X_train,
            epochs=epochs,
            batch_size=batch_size,
            validation_split=validation_split,
            callbacks=[early_stopping],
            verbose=1
        )

    def predict(self, X):
        reconstructed = self.autoencoder.predict(X)
        mse = np.mean(np.power(X - reconstructed, 2), axis=1)
        return mse

class ZeroDayDetector:
    def __init__(self, data_path=None, test_size=0.2, random_state=42):
        self.data_path = data_path
        self.test_size = test_size
        self.random_state = random_state
        self.models = {
            'one_class_svm': OneClassSVM(kernel='rbf', nu=0.1),
            'isolation_forest': IsolationForest(contamination=0.1, random_state=random_state),
            'kmeans': KMeans(n_clusters=2, random_state=random_state),
            'dbscan': DBSCAN(eps=0.5, min_samples=5),
            'autoencoder': None  # Will be initialized during training
        }
        self.results = {}
        self.X_train = None
        self.X_test = None
        self.y_train = None
        self.y_test = None
        self.scaler = StandardScaler()
        self.label_encoder = LabelEncoder()
        self.robust_scaler = RobustScaler()  # For outlier-resistant scaling
        self.pca = PCA(n_components=0.95)  # Preserve 95% of variance

    def preprocess_data(self, df):
        """Enhanced Data Preprocessing Step"""
        print("Step 1: Enhanced Data Preprocessing")

        # 1. Check for and handle duplicate rows
        duplicate_count = df.duplicated().sum()
        print(f"\nDuplicate rows found: {duplicate_count}")
        df = df.drop_duplicates()
        print(f"After removing duplicates: {len(df)} rows")

        # 2. Check for missing values
        missing_values = df.isnull().sum()
        print(f"\nColumns with missing values: {missing_values[missing_values > 0].index.tolist()}")

        # 3. Handle infinite values
        inf_count = np.isinf(df.select_dtypes(include=[np.number])).sum().sum()
        print(f"Infinite values found: {inf_count}")
        df = df.replace([np.inf, -np.inf], np.nan)

        # 4. Handle missing values (after replacing infinities with NaN)
        df = df.dropna()
        print(f"After handling missing/infinite values: {len(df)} rows")

        # 5. Check for and visualize class imbalance
        print("\nClass distribution before balancing:")
        class_dist = df[' Label'].value_counts()
        print(class_dist)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=class_dist.index, y=class_dist.values)
        plt.title('Class Distribution - Original Data')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # 6. Detect outliers
        self._detect_outliers(df)

        # 7. Separate features and labels
        X = df.drop(' Label', axis=1)
        y = df[' Label']

        # 8. Encode labels
        y_encoded = self.label_encoder.fit_transform(y)

        # 9. Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X, y_encoded, test_size=self.test_size, random_state=self.random_state, stratify=y_encoded
        )

        # 10. Handle class imbalance with SMOTE
        print("\nApplying SMOTE to balance classes...")
        smote = SMOTE(random_state=self.random_state)
        X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

        print("Class distribution after SMOTE:")
        balanced_dist = pd.Series(y_train_resampled).value_counts()
        print(balanced_dist)

        # 11. Scale features (after SMOTE to prevent data leakage)
        print("\nScaling features...")
        X_train_scaled = self.scaler.fit_transform(X_train_resampled)
        X_test_scaled = self.scaler.transform(X_test)

        self.X_train = X_train_scaled
        self.X_test = X_test_scaled
        self.y_train = y_train_resampled
        self.y_test = y_test

        return X_train_scaled, X_test_scaled, y_train_resampled, y_test, X.columns

    def _detect_outliers(self, df):
        """Detect and visualize outliers in the dataset"""
        print("\nDetecting outliers...")

        # Select numeric columns only
        numeric_df = df.select_dtypes(include=[np.number]).drop(' Label', axis=1, errors='ignore')

        # Create boxplots for selected features (top 5 by variance)
        variances = numeric_df.var().sort_values(ascending=False)
        top_features = variances.index[:5].tolist()

        plt.figure(figsize=(15, 10))
        for i, feature in enumerate(top_features):
            plt.subplot(2, 3, i+1)
            sns.boxplot(x=df[' Label'], y=df[feature])
            plt.title(f'Outliers in {feature}')
            plt.xticks(rotation=45)

        plt.tight_layout()
        plt.show()

        # Use IsolationForest to detect outliers
        iso_forest = IsolationForest(contamination=0.05, random_state=self.random_state)
        outliers = iso_forest.fit_predict(numeric_df)
        outlier_count = (outliers == -1).sum()

        print(f"Potential outliers detected: {outlier_count} ({outlier_count/len(df)*100:.2f}%)")

        # Visualize outliers using 2D PCA
        pca = PCA(n_components=2)
        pca_result = pca.fit_transform(numeric_df)

        plt.figure(figsize=(10, 6))
        plt.scatter(pca_result[outliers == 1, 0], pca_result[outliers == 1, 1],
                   c='blue', s=5, label='Normal')
        plt.scatter(pca_result[outliers == -1, 0], pca_result[outliers == -1, 1],
                   c='red', s=5, label='Outlier')
        plt.title('PCA visualization of outliers')
        plt.legend()
        plt.show()

        return outliers

    def feature_engineering(self, X_train_scaled, X_test_scaled, feature_names):
        """Enhanced Feature Selection and Engineering Step"""
        print("\nStep 2: Enhanced Feature Engineering")

        # 1. Check feature correlations
        print("Analyzing feature correlations...")

        # Convert scaled arrays back to DataFrames for correlation analysis
        X_train_df = pd.DataFrame(X_train_scaled, columns=feature_names)

        # Plot correlation matrix
        plt.figure(figsize=(12, 10))
        corr_matrix = X_train_df.corr()
        mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
        sns.heatmap(corr_matrix, mask=mask, cmap='coolwarm', vmax=1, vmin=-1, center=0,
                   square=True, linewidths=.5, annot=False)
        plt.title('Feature Correlation Matrix')
        plt.tight_layout()
        plt.show()

        # 2. Apply PCA
        print("Applying PCA...")
        X_train_pca = self.pca.fit_transform(X_train_scaled)
        X_test_pca = self.pca.transform(X_test_scaled)

        # Get number of components and explained variance
        n_components = self.pca.n_components_
        explained_variance = self.pca.explained_variance_ratio_.sum()
        print(f"Number of PCA components: {n_components}")
        print(f"Explained variance: {explained_variance:.2%}")

        # 3. Plot explained variance ratio
        plt.figure(figsize=(10, 6))
        plt.plot(np.cumsum(self.pca.explained_variance_ratio_))
        plt.xlabel('Number of Components')
        plt.ylabel('Cumulative Explained Variance')
        plt.title('Explained Variance by PCA Components')
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        # 4. Get feature importance
        components_df = pd.DataFrame(
            self.pca.components_.T,
            columns=[f'PC{i+1}' for i in range(self.pca.components_.shape[0])],
            index=feature_names
        )

        # Calculate and sort feature importances
        feature_importance = np.abs(self.pca.components_).mean(axis=0)
        feature_importance = pd.Series(feature_importance, index=feature_names).sort_values(ascending=False)

        # Plot top 15 most important features
        plt.figure(figsize=(12, 6))
        feature_importance.head(15).plot(kind='bar')
        plt.title('Top 15 Most Important Features')
        plt.tight_layout()
        plt.show()

        # 5. Plot feature contribution heatmap
        plt.figure(figsize=(12, 8))
        sns.heatmap(components_df.loc[feature_importance.index[:15]].T, cmap='viridis', annot=False)
        plt.title('Top 15 Features Contribution to Principal Components')
        plt.tight_layout()
        plt.show()

        return X_train_pca, X_test_pca, feature_importance

    def train_models(self):
        """Train all models and store their results."""
        print("Training models...")

        # Initialize and train autoencoder
        input_dim = self.X_train.shape[1]
        self.models['autoencoder'] = Autoencoder(input_dim=input_dim)
        self.models['autoencoder'].fit(self.X_train)

        for model_name, model in self.models.items():
            print(f"\nTraining {model_name}...")
            start_time = time.time()

            if model_name == 'autoencoder':
                # Get reconstruction error and set threshold
                train_scores = model.predict(self.X_train)
                threshold = np.percentile(train_scores, 90)  # Adjust percentile as needed
                train_predictions = (train_scores > threshold).astype(int)

                test_scores = model.predict(self.X_test)
                test_predictions = (test_scores > threshold).astype(int)
            else:
                # Train and predict with other models as before
                if model_name in ['one_class_svm', 'isolation_forest']:
                    model.fit(self.X_train)
                    train_predictions = model.predict(self.X_train)
                    test_predictions = model.predict(self.X_test)
                    # Convert predictions from {1, -1} to {0, 1}
                    train_predictions = (train_predictions == -1).astype(int)
                    test_predictions = (test_predictions == -1).astype(int)
                else:  # KMeans and DBSCAN
                    model.fit(self.X_train)
                    train_predictions = model.labels_
                    test_predictions = model.fit_predict(self.X_test)

                    if model_name == 'dbscan':
                        # Convert noise labels (-1) to new cluster labels
                        train_predictions = self._process_dbscan_labels(train_predictions)
                        test_predictions = self._process_dbscan_labels(test_predictions)

                    # Map clusters to actual class labels
                    train_predictions = self._map_clusters_to_labels(train_predictions, self.y_train)
                    test_predictions = self._map_clusters_to_labels(test_predictions, self.y_test)

            # Calculate metrics
            train_metrics = self._calculate_metrics(self.y_train, train_predictions)
            test_metrics = self._calculate_metrics(self.y_test, test_predictions)

            # Store results
            self.results[model_name] = {
                'train_predictions': train_predictions,
                'test_predictions': test_predictions,
                'train_metrics': train_metrics,
                'test_metrics': test_metrics,
                'training_time': time.time() - start_time
            }

            print(f"{model_name} training completed in {self.results[model_name]['training_time']:.2f} seconds")
            print(f"Train metrics: {train_metrics}")
            print(f"Test metrics: {test_metrics}")

    def visualize_results(self, X, y, predictions, title):
        """Visualization of Results"""
        from sklearn.manifold import TSNE

        # Apply t-SNE
        tsne = TSNE(n_components=2, random_state=42)
        X_tsne = tsne.fit_transform(X)

        # Plot results
        plt.figure(figsize=(12, 4))

        # Original labels
        plt.subplot(1, 2, 1)
        scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
        plt.title(f'{title} - True Labels')
        plt.colorbar(scatter)

        # Predicted labels
        plt.subplot(1, 2, 2)
        scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=predictions, cmap='viridis')
        plt.title(f'{title} - Predicted Labels')
        plt.colorbar(scatter)

        plt.tight_layout()
        plt.show()

    def _process_dbscan_labels(self, labels):
        """Convert DBSCAN noise labels (-1) to new cluster labels."""
        max_label = max(np.max(labels), 0)  # Handle case when all points are noise (-1)
        return np.where(labels == -1, max_label + 1, labels)

    def _map_clusters_to_labels(self, cluster_labels, true_labels):
        """Map cluster numbers to actual class labels based on majority class."""
        cluster_to_label = {}
        for cluster in np.unique(cluster_labels):
            mask = cluster_labels == cluster
            if np.sum(mask) > 0:
                most_common_label = Counter(true_labels[mask]).most_common(1)[0][0]
                cluster_to_label[cluster] = most_common_label

        return np.array([cluster_to_label.get(c, 0) for c in cluster_labels])

    def _calculate_metrics(self, y_true, y_pred):
        """Calculate classification metrics."""
        return {
            'accuracy': accuracy_score(y_true, y_pred),
            'precision': precision_score(y_true, y_pred, average='weighted', zero_division=0),
            'recall': recall_score(y_true, y_pred, average='weighted', zero_division=0),
            'f1': f1_score(y_true, y_pred, average='weighted', zero_division=0),
            'confusion_matrix': confusion_matrix(y_true, y_pred).tolist()
        }

def main():
    # Start timing
    start_time = time.time()

    # Initialize detector
    detector = ZeroDayDetector()

    # Load data with error handling
    print("Loading data...")
    try:
        # Try reading with default settings first
        df = pd.read_csv('combined_friday_samples2.csv')
    except pd.errors.ParserError:
        print("Error with default parsing, attempting with error handling...")
        # If that fails, try with error handling options
        df = pd.read_csv('combined_friday_samples2.csv',
                        on_bad_lines='skip',  # Skip problematic rows
                        warn_bad_lines=True,   # Warn about skipped rows
                        encoding='utf-8')      # Specify encoding
        print(f"Successfully loaded {len(df)} rows after skipping problematic ones")

    # Preprocess data
    X_train_scaled, X_test_scaled, y_train, y_test, feature_names = detector.preprocess_data(df)

    # Feature engineering
    X_train_pca, X_test_pca, feature_importance = detector.feature_engineering(X_train_scaled, X_test_scaled, feature_names)

    # Train only autoencoder
    print("\nTraining Autoencoder model...")
    input_dim = X_train_scaled.shape[1]
    autoencoder = Autoencoder(input_dim=input_dim)
    autoencoder.fit(X_train_scaled)

    # Get autoencoder predictions
    train_scores = autoencoder.predict(X_train_scaled)
    threshold = np.percentile(train_scores, 90)  # Adjust percentile as needed
    train_predictions = (train_scores > threshold).astype(int)

    test_scores = autoencoder.predict(X_test_scaled)
    test_predictions = (test_scores > threshold).astype(int)

    # Calculate metrics for autoencoder
    train_metrics = detector._calculate_metrics(y_train, train_predictions)
    test_metrics = detector._calculate_metrics(y_test, test_predictions)

    print("\nAutoencoder Performance Summary:")
    print(f"Train metrics: {train_metrics}")
    print(f"Test metrics: {test_metrics}")

    # Visualize autoencoder results
    detector.visualize_results(X_test_pca, y_test, test_predictions, "Autoencoder")

    # Print execution time
    end_time = time.time()
    print(f"\nTotal execution time: {(end_time - start_time)/60:.2f} minutes")

    # Print most important features
    print("\nTop 10 most important features:")
    print(feature_importance.head(10))

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import os
import numpy as np
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler, LabelEncoder, RobustScaler
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE, ADASYN
from collections import Counter
import time
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix, classification_report, silhouette_score
from sklearn.metrics import classification_report, confusion_matrix, f1_score, precision_score, recall_score
from sklearn.feature_selection import VarianceThreshold

class ZeroDayDetector:
    def __init__(self):
        self.scaler = StandardScaler()
        self.robust_scaler = RobustScaler()  # For outlier-resistant scaling
        self.label_encoder = LabelEncoder()
        self.pca = PCA(n_components=0.95)  # Preserve 95% of variance

    def preprocess_data(self, df):
        """Enhanced Data Preprocessing Step"""
        print("Step 1: Enhanced Data Preprocessing")

        # 1. Check for and handle duplicate rows
        duplicate_count = df.duplicated().sum()
        print(f"\nDuplicate rows found: {duplicate_count}")
        df = df.drop_duplicates()
        print(f"After removing duplicates: {len(df)} rows")

        # 2. Check for missing values
        missing_values = df.isnull().sum()
        print(f"\nColumns with missing values: {missing_values[missing_values > 0].index.tolist()}")

        # 3. Handle infinite values
        inf_count = np.isinf(df.select_dtypes(include=[np.number])).sum().sum()
        print(f"Infinite values found: {inf_count}")
        df = df.replace([np.inf, -np.inf], np.nan)

        # 4. Handle missing values (after replacing infinities with NaN)
        df = df.dropna()
        print(f"After handling missing/infinite values: {len(df)} rows")

        # 5. Check for and visualize class imbalance
        print("\nClass distribution before balancing:")
        class_dist = df[' Label'].value_counts()
        print(class_dist)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=class_dist.index, y=class_dist.values)
        plt.title('Class Distribution - Original Data')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # 6. Detect outliers
        self._detect_outliers(df)

        # 7. Separate features and labels
        X = df.drop(' Label', axis=1)
        y = df[' Label']

        # 8. Encode labels
        y_encoded = self.label_encoder.fit_transform(y)

        # 9. Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
        )

        # 10. Handle class imbalance with SMOTE
        print("\nApplying SMOTE to balance classes...")
        smote = SMOTE(random_state=42)
        X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

        print("Class distribution after SMOTE:")
        balanced_dist = pd.Series(y_train_resampled).value_counts()
        print(balanced_dist)

        # 11. Scale features (after SMOTE to prevent data leakage)
        print("\nScaling features...")
        X_train_scaled = self.scaler.fit_transform(X_train_resampled)
        X_test_scaled = self.scaler.transform(X_test)

        return X_train_scaled, X_test_scaled, y_train_resampled, y_test, X.columns

    def _detect_outliers(self, df):
        """Detect and visualize outliers in the dataset"""
        print("\nDetecting outliers...")

        # Select numeric columns only
        numeric_df = df.select_dtypes(include=[np.number]).drop(' Label', axis=1, errors='ignore')

        # Create boxplots for selected features (top 5 by variance)
        variances = numeric_df.var().sort_values(ascending=False)
        top_features = variances.index[:5].tolist()

        plt.figure(figsize=(15, 10))
        for i, feature in enumerate(top_features):
            plt.subplot(2, 3, i+1)
            sns.boxplot(x=df[' Label'], y=df[feature])
            plt.title(f'Outliers in {feature}')
            plt.xticks(rotation=45)

        plt.tight_layout()
        plt.show()

        # Use IsolationForest to detect outliers
        iso_forest = IsolationForest(contamination=0.05, random_state=42)
        outliers = iso_forest.fit_predict(numeric_df)
        outlier_count = (outliers == -1).sum()

        print(f"Potential outliers detected: {outlier_count} ({outlier_count/len(df)*100:.2f}%)")

        # Visualize outliers using 2D PCA
        pca = PCA(n_components=2)
        pca_result = pca.fit_transform(numeric_df)

        plt.figure(figsize=(10, 6))
        plt.scatter(pca_result[outliers == 1, 0], pca_result[outliers == 1, 1],
                   c='blue', s=5, label='Normal')
        plt.scatter(pca_result[outliers == -1, 0], pca_result[outliers == -1, 1],
                   c='red', s=5, label='Outlier')
        plt.title('PCA visualization of outliers')
        plt.legend()
        plt.show()

        return outliers

    def feature_engineering(self, X_train_scaled, X_test_scaled, feature_names):
        """Enhanced Feature Selection and Engineering Step"""
        print("\nStep 2: Enhanced Feature Engineering")

        # 1. Check feature correlations
        print("Analyzing feature correlations...")

        # Convert scaled arrays back to DataFrames for correlation analysis
        X_train_df = pd.DataFrame(X_train_scaled, columns=feature_names)

        # Plot correlation matrix
        plt.figure(figsize=(12, 10))
        corr_matrix = X_train_df.corr()
        mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
        sns.heatmap(corr_matrix, mask=mask, cmap='coolwarm', vmax=1, vmin=-1, center=0,
                   square=True, linewidths=.5, annot=False)
        plt.title('Feature Correlation Matrix')
        plt.tight_layout()
        plt.show()

        # 2. Apply PCA
        print("Applying PCA...")
        X_train_pca = self.pca.fit_transform(X_train_scaled)
        X_test_pca = self.pca.transform(X_test_scaled)

        # Get number of components and explained variance
        n_components = self.pca.n_components_
        explained_variance = self.pca.explained_variance_ratio_.sum()
        print(f"Number of PCA components: {n_components}")
        print(f"Explained variance: {explained_variance:.2%}")

        # 3. Plot explained variance ratio
        plt.figure(figsize=(10, 6))
        plt.plot(np.cumsum(self.pca.explained_variance_ratio_))
        plt.xlabel('Number of Components')
        plt.ylabel('Cumulative Explained Variance')
        plt.title('Explained Variance by PCA Components')
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        # 4. Get feature importance
        components_df = pd.DataFrame(
            self.pca.components_.T,
            columns=[f'PC{i+1}' for i in range(self.pca.components_.shape[0])],
            index=feature_names
        )

        # Calculate and sort feature importances
        feature_importance = np.abs(self.pca.components_).mean(axis=0)
        feature_importance = pd.Series(feature_importance, index=feature_names).sort_values(ascending=False)

        # Plot top 15 most important features
        plt.figure(figsize=(12, 6))
        feature_importance.head(15).plot(kind='bar')
        plt.title('Top 15 Most Important Features')
        plt.tight_layout()
        plt.show()

        # 5. Plot feature contribution heatmap
        plt.figure(figsize=(12, 8))
        sns.heatmap(components_df.loc[feature_importance.index[:15]].T, cmap='viridis', annot=False)
        plt.title('Top 15 Features Contribution to Principal Components')
        plt.tight_layout()
        plt.show()

        return X_train_pca, X_test_pca, feature_importance
    def train_models(self, X_train, X_test, y_train, y_test):
        """Model Training and Evaluation"""
        print("\nStep 3: Model Training and Evaluation")

        # Convert multi-class labels to binary (0 = normal/benign, 1 = anomaly/attack)
        # Assuming class 0 is benign/normal and all other classes (1,2,3) are different attack types
        print("\nConverting multi-class labels to binary (benign vs attack) for anomaly detection...")
        y_train_binary = np.where(y_train == 0, 0, 1)  # 0 remains 0 (benign), all other classes become 1 (attack)
        y_test_binary = np.where(y_test == 0, 0, 1)

        print(f"Binary class distribution in training set: {np.bincount(y_train_binary)}")
        print(f"Binary class distribution in test set: {np.bincount(y_test_binary)}")

        # For unsupervised training of anomaly detectors, we'll use only benign samples
        # This simulates a real-world scenario where only normal behavior is known during training
        print("\nFor unsupervised anomaly detection, training only on benign (class 0) samples...")
        X_train_benign = X_train[y_train == 0]
        print(f"Training on {len(X_train_benign)} benign samples")

        # Try different contamination rates for better performance
        contamination_rates = [0.01, 0.05, 0.1, 0.15, 0.2]
        best_models = {}
        best_f1_scores = {'One-Class SVM': 0, 'Isolation Forest': 0}

        print("\nTuning anomaly detection models for better performance...")

        # Feature selection using variance threshold to remove low-variance features
        print("\nApplying variance threshold for feature selection...")
        selector = VarianceThreshold(threshold=0.01)  # Remove features with low variance
        X_train_selected = selector.fit_transform(X_train)
        X_test_selected = selector.transform(X_test)
        X_train_benign_selected = selector.transform(X_train_benign)
        print(f"Features reduced from {X_train.shape[1]} to {X_train_selected.shape[1]}")

        # Try different parameters for One-Class SVM
        print("\nTuning One-Class SVM...")
        svm_params = [
            {'kernel': 'rbf', 'nu': 0.05, 'gamma': 'scale'},
            {'kernel': 'rbf', 'nu': 0.1, 'gamma': 'scale'},
            {'kernel': 'rbf', 'nu': 0.05, 'gamma': 'auto'},
            {'kernel': 'linear', 'nu': 0.05},
            {'kernel': 'sigmoid', 'nu': 0.05, 'gamma': 'scale'}
        ]

        best_svm = None
        best_svm_f1 = 0
        best_svm_params = None

        for params in svm_params:
            print(f"Testing One-Class SVM with params: {params}")
            svm = OneClassSVM(**params)
            svm.fit(X_train_benign_selected)

            # Predict on test set
            test_pred_raw = svm.predict(X_test_selected)
            test_pred = np.where(test_pred_raw == -1, 1, 0)

            # Calculate F1 score
            f1 = f1_score(y_test_binary, test_pred)
            print(f"F1 score: {f1:.4f}")

            if f1 > best_svm_f1:
                best_svm_f1 = f1
                best_svm = svm
                best_svm_params = params

        print(f"\nBest One-Class SVM parameters: {best_svm_params}, F1: {best_svm_f1:.4f}")

        # Try different parameters for Isolation Forest
        print("\nTuning Isolation Forest...")
        if_params = [
            {'contamination': 0.05, 'max_features': 1.0, 'n_estimators': 100},
            {'contamination': 0.1, 'max_features': 0.8, 'n_estimators': 100},
            {'contamination': 0.15, 'max_features': 0.8, 'n_estimators': 200},
            {'contamination': 0.2, 'max_features': 0.7, 'n_estimators': 100},
            {'contamination': 'auto', 'max_features': 0.8, 'n_estimators': 200}
        ]

        best_if = None
        best_if_f1 = 0
        best_if_params = None

        for params in if_params:
            print(f"Testing Isolation Forest with params: {params}")
            iso_forest = IsolationForest(**params, random_state=42, n_jobs=-1)
            iso_forest.fit(X_train_benign_selected)

            # Predict on test set
            test_pred_raw = iso_forest.predict(X_test_selected)
            test_pred = np.where(test_pred_raw == -1, 1, 0)

            # Calculate F1 score
            f1 = f1_score(y_test_binary, test_pred)
            print(f"F1 score: {f1:.4f}")

            if f1 > best_if_f1:
                best_if_f1 = f1
                best_if = iso_forest
                best_if_params = params

        print(f"\nBest Isolation Forest parameters: {best_if_params}, F1: {best_if_f1:.4f}")

        # Use the best models found
        models = {
            'One-Class SVM': best_svm if best_svm is not None else OneClassSVM(kernel='rbf', nu=0.05, gamma='scale'),
            'Isolation Forest': best_if if best_if is not None else IsolationForest(contamination=0.05, random_state=42, n_jobs=-1),
            #'K-Means': KMeans(n_clusters=len(np.unique(y_train)), random_state=42, n_init=10),
        }

        results = {}

        for name, model in models.items():
            print(f"\nTraining {name}...")

            if name in ['One-Class SVM', 'Isolation Forest']:
                # Models are already trained during parameter tuning
                if name == 'One-Class SVM' and best_svm is None:
                    model.fit(X_train_benign_selected)
                elif name == 'Isolation Forest' and best_if is None:
                    model.fit(X_train_benign_selected)

                # Get predictions (1 = normal, -1 = anomaly for these algorithms)
                train_pred_raw = model.predict(X_train_selected)
                test_pred_raw = model.predict(X_test_selected)

                # Convert to binary classification format (0 = benign, 1 = attack)
                # -1 (anomaly) becomes 1 (attack), 1 (normal) becomes 0 (benign)
                train_pred = np.where(train_pred_raw == -1, 1, 0)
                test_pred = np.where(test_pred_raw == -1, 1, 0)

                # Evaluate against binary ground truth
                train_report = classification_report(y_train_binary, train_pred, zero_division=0)
                test_report = classification_report(y_test_binary, test_pred, zero_division=0)

                # Additional metrics
                test_f1 = f1_score(y_test_binary, test_pred, zero_division=0)
                test_precision = precision_score(y_test_binary, test_pred, zero_division=0)
                test_recall = recall_score(y_test_binary, test_pred, zero_division=0)

                # Plot confusion matrices for binary classification
                plt.figure(figsize=(12, 5))

                plt.subplot(1, 2, 1)
                cm_train = confusion_matrix(y_train_binary, train_pred)
                sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues',
                           xticklabels=['Benign', 'Attack'], yticklabels=['Benign', 'Attack'])
                plt.xlabel('Predicted')
                plt.ylabel('True')
                plt.title(f'{name} - Training Confusion Matrix (Binary)')

                plt.subplot(1, 2, 2)
                cm_test = confusion_matrix(y_test_binary, test_pred)
                sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues',
                           xticklabels=['Benign', 'Attack'], yticklabels=['Benign', 'Attack'])
                plt.xlabel('Predicted')
                plt.ylabel('True')
                plt.title(f'{name} - Testing Confusion Matrix (Binary)')

                plt.tight_layout()
                plt.show()

                # Also show detailed attack type detection performance
                print("\nDetailed attack type detection performance:")
                for attack_type in range(1, 4):  # Assuming attack types are 1, 2, 3
                    if attack_type in np.unique(y_test):
                        attack_indices = (y_test == attack_type)
                        if np.sum(attack_indices) > 0:
                            attack_detection = test_pred[attack_indices]
                            detection_rate = np.sum(attack_detection) / len(attack_detection) * 100
                            print(f"Attack type {attack_type}: detected {np.sum(attack_detection)}/{len(attack_detection)} " +
                                 f"({detection_rate:.2f}%)")

            else:  # K-Means
                # For K-Means, we'll use the improved approach with cluster alignment
                model.fit(X_train)

                # Get raw cluster predictions
                train_pred_raw = model.predict(X_train)
                test_pred_raw = model.predict(X_test)

                # Align clusters with true labels
                train_pred = self._align_cluster_labels(train_pred_raw, y_train)
                test_pred = self._align_cluster_labels(test_pred_raw, y_test, from_train=False,
                                                      train_clusters=train_pred_raw, train_labels=y_train)

                # Calculate metrics for multi-class
                train_report = classification_report(y_train, train_pred, zero_division=0)
                test_report = classification_report(y_test, test_pred, zero_division=0)

                # Additional metrics
                test_f1 = f1_score(y_test, test_pred, average='weighted', zero_division=0)
                test_precision = precision_score(y_test, test_pred, average='weighted', zero_division=0)
                test_recall = recall_score(y_test, test_pred, average='weighted', zero_division=0)

                # Plot confusion matrices for multi-class
                plt.figure(figsize=(12, 5))

                plt.subplot(1, 2, 1)
                cm_train = confusion_matrix(y_train, train_pred)
                sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues')
                plt.title(f'{name} - Training Confusion Matrix')

                plt.subplot(1, 2, 2)
                cm_test = confusion_matrix(y_test, test_pred)
                sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues')
                plt.title(f'{name} - Testing Confusion Matrix')

                plt.tight_layout()
                plt.show()

            # Store results
            results[name] = {
                'train_pred': train_pred,
                'test_pred': test_pred,
                'train_report': train_report,
                'test_report': test_report,
                'test_f1': test_f1,
                'test_precision': test_precision,
                'test_recall': test_recall
            }

            print(f"\n{name} Training Report:")
            print(train_report)
            print(f"\n{name} Testing Report:")
            print(test_report)
            print(f"F1 Score (Test): {test_f1:.4f}")
            print(f"Precision (Test): {test_precision:.4f}")
            print(f"Recall (Test): {test_recall:.4f}")

        return results

    def _align_cluster_labels(self, cluster_labels, true_labels, from_train=True, train_clusters=None, train_labels=None):
        """
        Align cluster labels with true labels to maximize accuracy.

        Parameters:
        -----------
        cluster_labels : array-like
            Predicted cluster labels
        true_labels : array-like
            True class labels
        from_train : bool
            Whether this is training data or test data
        train_clusters : array-like
            Training cluster labels (only used when from_train=False)
        train_labels : array-like
            Training true labels (only used when from_train=False)

        Returns:
        --------
        aligned_labels : array-like
            Aligned cluster labels
        """
        if from_train:
            # For training data, find the best mapping
            n_clusters = len(np.unique(cluster_labels))
            n_classes = len(np.unique(true_labels))

            # Create a mapping from clusters to most common class
            mapping = {}
            for i in range(n_clusters):
                mask = (cluster_labels == i)
                if np.sum(mask) > 0:  # Ensure the cluster is not empty
                    # Find most common class in this cluster
                    most_common_class = np.bincount(true_labels[mask]).argmax()
                    mapping[i] = most_common_class
                else:
                    # Assign a default class if cluster is empty
                    mapping[i] = 0

            # Apply mapping
            aligned_labels = np.array([mapping[label] for label in cluster_labels])
            return aligned_labels

        else:
            # For test data, use the mapping from training data
            assert train_clusters is not None and train_labels is not None

            # Create mapping from training data
            mapping = {}
            for cluster in np.unique(train_clusters):
                mask = (train_clusters == cluster)
                if np.sum(mask) > 0:
                    most_common_class = np.bincount(train_labels[mask]).argmax()
                    mapping[cluster] = most_common_class
                else:
                    mapping[cluster] = 0

            # Apply mapping to test data
            aligned_labels = np.array([mapping.get(label, 0) for label in cluster_labels])
            return aligned_labels

    def visualize_results(self, X, y, predictions, title):
        """Visualization of Results"""
        from sklearn.manifold import TSNE

        # Apply t-SNE
        tsne = TSNE(n_components=2, random_state=42)
        X_tsne = tsne.fit_transform(X)

        # Plot results
        plt.figure(figsize=(12, 4))

        # Original labels
        plt.subplot(1, 2, 1)
        scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
        plt.title(f'{title} - True Labels')
        plt.colorbar(scatter)

        # Predicted labels
        plt.subplot(1, 2, 2)
        scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=predictions, cmap='viridis')
        plt.title(f'{title} - Predicted Labels')
        plt.colorbar(scatter)

        plt.tight_layout()
        plt.show()

def main():
    # Start timing
    start_time = time.time()

    # Initialize detector
    detector = ZeroDayDetector()

    # Load data
    print("Loading data...")
    df = pd.read_csv('combined_friday_samples2.csv')

    # Preprocess data
    X_train_scaled, X_test_scaled, y_train, y_test, feature_names = detector.preprocess_data(df)

    # Feature engineering
    X_train_pca, X_test_pca, feature_importance = detector.feature_engineering(X_train_scaled, X_test_scaled, feature_names)

    # Train and evaluate models
    results = detector.train_models(X_train_pca, X_test_pca, y_train, y_test)

    # Print execution time
    end_time = time.time()
    print(f"\nTotal execution time: {(end_time - start_time)/60:.2f} minutes")

    # Print most important features
    print("\nTop 10 most important features:")
    print(feature_importance.head(10))

if __name__ == "__main__":
    main()


```
